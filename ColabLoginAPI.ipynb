# **LOGIN API**
Bu aşamaya kadar "Flask" ve "Ngrok" paketleri yüklendi. Aşağıdaki örnek kod ile basit bir *login* metodu yazılmıştır.

```python
from flask_ngrok import run_with_ngrok
from flask import Flask, request, jsonify

app = Flask(__name__)
run_with_ngrok(app)

email = "admin@admin.com"
password = "1234"

admin_info = {"name": "Admin", "email": email, "age": 20, "permission": "all"}

@app.route('/login',methods = ['POST'])
def login():
  if request.method == 'POST':
    if request.is_json:
      data = request.get_json();
      if data['email'] == email and data['password'] == password:
        return admin_info

app.run()
```
